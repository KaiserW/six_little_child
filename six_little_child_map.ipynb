```python
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
mblog_df = pd.read_csv('six_little_child_weibo_181122.csv')
```

```python
mblog_df.head()
```

```python
six_weibo_list = [BeautifulSoup(mblog, 'html5lib').text for mblog in mblog_df['text']]
```

```python
import jieba.posseg
```

```python
for weibo in six_weibo_list[10:20]:
    words = jieba.posseg.cut(weibo)
    for word, flag in words:
        if flag == 'ns':
            print('词语：{} -- 词性：{}'.format(word, flag))
```

```python
places_visited = []
for weibo in tqdm(six_weibo_list):
    
    words = jieba.posseg.cut( weibo)
    place_names = []
    
    for word, flag in words:
        if flag == 'ns' and word not in place_names:
            place_names.append(word)
    
    places_visited.append(' '.join(place_names))
```

```python
mblog_df['places'] = places_visited
```
